```python
import seaborn as sns
sns.set(style="whitegrid")
```

```python
sns.countplot(data=data2, x='Attrition')
plt.title("Employee Attrition Count")
plt.xlabel("Attrition (1 = Yes, 0 = No)")
plt.ylabel("Number of Employees")
plt.show()
```

```python
data2.count().sum()
```

```python
sns.countplot(data=data2, x='Department', hue='Attrition')
plt.title("Attrition by Department")
plt.xticks(rotation=30)
plt.show()
```

```python
sns.countplot(data=data2, x='Gender', hue='Attrition')
plt.title("Attrition by Gender")
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.countplot(data=data2, x='JobRole', hue='Attrition')
plt.title("Attrition by Job Role")
plt.xticks(rotation=30)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=data2, x='Age', hue='Attrition', multiple='stack', bins=30)
plt.title("Age Distribution by Attrition")
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.boxplot(x='Attrition', y='MonthlyIncome', data=data2)
plt.title("Income Distribution by Attrition")
plt.show()
```
